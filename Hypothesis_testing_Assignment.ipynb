**6.2 ESTIMATION AND CONFIDENCE INTERVAL**

```python
# Part (a): 99% Confidence Interval Using Sample Standard Deviation

import numpy as np
from scipy.stats import norm
```

```python
# Creating a NumPy array containing the sample data values
data = np.array([1.13, 1.55, 1.43, 0.92, 1.25, 1.36, 1.32, 0.85, 1.07, 1.48, 1.20, 1.33, 1.18, 1.22, 1.29])

# Finding the total number of observations in the dataset
n = len(data)

# Calculating the sample mean
mean = np.mean(data)

# Calculating the sample standard deviation
sample_std = np.std(data)

# Standard error of the mean (std / sqrt(sample size))
fails = sample_std / np.sqrt(sample_size)

# Getting the z-critical value for a 99% confidence interval (0.5% in upper tail)
z_critical = norm.ppf(0.995)

# Lower bound of the confidence interval
lower = mean - z_critical * fails

# Upper bound of the confidence interval
upper = mean + z_critical * fails
```

```python
# Part (b): 99% Confidence Interval Using Known Population Standard Deviation
```

```python
from scipy.stats import norm
```

```python
# Population standard deviation is given as 0.2
pop_std_dev = 0.2

# Z-critical value for a 99% confidence interval (0.5% in each tail)
z_critical = norm.ppf(0.995)

# Calculating the lower bound of the confidence interval using population standard deviation
lower = mean - z_critical * (pop_std_dev / np.sqrt(n))

# Calculating the upper bound of the confidence interval using population standard deviation
upper = mean + z_critical * (pop_std_dev / np.sqrt(n))
```
